```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
df=pd.DataFrame(columns=['Job Title','Company Name','Link'])
```

```python
def get_data(title,location):
    
    url='https://www.linkedin.com/jobs/search?keywords='+str(title)+'&location='+str(location)
    response=requests.get(url)
    soup=BeautifulSoup(response.content,'html.parser')
    for d in soup.find_all('div', attrs={'class':"base-search-card"}):
        title=d.find('h3',attrs={'class':"base-search-card__title"})
        company_name=d.find('h4',attrs={'class':"base-search-card__subtitle"})
        link=d.find('a',attrs={'class':"base-card__full-link"})
        data=[]
        if title is not None:
            data.append(title.text)
        else:
            data.append('nan')
        if company_name is not None:
            data.append(company_name.text)
        else:
            data.append('nan')
        if link is not None:
            data.append(link['href'])
        else:
            data.append('nan')
        df.loc[len(df)]=data
    return df
   
```

```python
get_data('software','delhi')
df
```
